```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris
```

```python
iris.target
```

```python
# total number of neurons in input layer - 4
# total number of neurons in hidden layer adjacent to input layer - 2/3 of input layer
# total number of hidden neurons in adjacent layer is not fixed, if complexity is less, reduce the neurons , if underfit then increase the neurons
```

```python
iris.data.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
# Sequential is used when we want our layers in a sqauence
# Dense is used to maintain the order
# relu is used to overcome vanishing gradient problem
model=Sequential([
    Dense(units=3,activation="relu",input_shape=(X_train.shape[1],)),# hidden layer
    Dense(units=3,activation="softmax",)
])
```

```python
model.summary()
```

```python
model.compile(optimizer="SGD",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
# optimizer --> stochastic gradient descent
```

```python
model.fit(X_train,y_train,epochs=1000, verbose=1)
```

```python
model.evaluate(X_test,y_test)
```

```python
model.predict(X_test)
```
